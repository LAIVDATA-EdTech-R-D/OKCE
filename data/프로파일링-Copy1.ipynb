```python
df.groupby(['Problem Hierarchy','Anon Student Id']).nunique()
```

```python
def func(x):
    d = {}
    d['Anon Student Id Count'] = x['Anon Student Id'].nunique()
    return pd.Series(d, index=['Anon Student Id Count'])
```

```python
df_return = df.groupby(['Problem Hierarchy', 'KC(SubSkills)']).apply(func)
df_return = df_return.sort_values('Anon Student Id Count', ascending=False) 
df_return
```
