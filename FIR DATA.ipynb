```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./algebra_ratio3000.csv')
```

```python
df = df.sort_values('Unnamed: 0')
df = df.reset_index(drop=True)
df
```

```python
# selector net에서 오는 데이터 example
target = "Calculate.total.in.proportion.with.fractions"
KC_list = ["Enter.given.part.in.proportion", "Enter.Calculated.value.of.rate", 
           'Enter.denominator.of.form.of.1', 'Enter.proportion.label.in.numerator']
```

```python
temp_list = KC_list
temp_list.append(target)
df = pd.concat([df['Unnamed: 0'],df[temp_list]],axis=1)
df.head()
```

```python
re_cols = list(df.columns)
re_cols.remove(target)
re_cols.append(target)
re_cols[-2:]
```

```python
df = df[re_cols]
df.head(1)
```

```python
stus = list(df['Unnamed: 0'].unique())
len(stus)
```

```python
index = []
skill = []
acc = []

for i in range(len(stus)):
    for j in range(len(re_cols[1:])):
        index.append(stus[i])
        skill.append(re_cols[j+1])
        acc.append(df.loc[i][j+1])
```

```python
pd.DataFrame({ 'student':index, 'KC':skill, 'acc':acc })
```
